```python
import datetime as dt
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn import metrics

raw_product_data = pd.read_csv("df.csv", index_col=0)
raw_product_data['Date'] = pd.to_datetime(raw_product_data['Date'])
raw_product_data['Date_ordinal']=raw_product_data['Date'].map(dt.datetime.toordinal)
# raw_product_data['Date']=raw_product_data['Date'].map(dt.datetime.toordinal)

raw_product_data= raw_product_data[raw_product_data['Price'] != 0]
```
